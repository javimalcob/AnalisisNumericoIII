```python
import numpy as np
```

```python
def f(x):
  return np.sin(np.pi * x)
```

```python
#parametros
h = 0.2
x0 = 0.4
```

```python
(f(x0 - 2*h) - 8*f(x0 - h) + 8*f(x0 + h) - f(x0 + 2*h))/ (12*h)
```

```python
#valor real
valor_real = np.pi * np.cos(np.pi * x0)
```

```python
h_vec = [0.2, 0.1, 0.05, 0.01, 0.005, 0.001, 0.0005, 0.0001, 0.00001]
```

```python
err_vec = []
for h in h_vec:
  valor_aprox = (f(x0 + h) - f(x0 - h))/ (2*h)
  err = valor_aprox - valor_real

  print(h ,' : ', valor_aprox, 'error: ', abs(err))
  err_vec.append(abs(err))

print(err_vec)
```

```python
err_vec
for i in range(len(err_vec)-1):
  p = np.log(err_vec[i]/err_vec[i + 1])/np.log(h_vec[i]/h_vec[i + 1])
  print(p)
```
